```python
# 参考リンク
# https://stackoverflow.com/questions/37435369/matplotlib-how-to-draw-a-rectangle-on-image

#枠が正しいか確認
import numpy as np
from PIL import *
from matplotlib import patches

im = np.array(Image.open('img/stop_sign-003.jpg'), dtype=np.uint8)

# Create figure and axes
fig,ax = plt.subplots(1)

# Display the image
ax.imshow(im)

# Create a Rectangle patch
rect = patches.Rectangle((193,134),227,201,linewidth=1,edgecolor='r',facecolor='none')
# Add the patch to the Axes
ax.add_patch(rect)

plt.show()
```
